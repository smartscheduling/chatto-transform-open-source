```python
from chatto_transform.datastores.mimic.chartevents_datastore import CharteventsDataStore
from chatto_transform.datastores.csv_datastore import CsvDataStore
from chatto_transform.schema.schema_base import Schema
from chatto_transform.schema.mimic.mimic_schema import chartevents_schema
```

### You'll need to change the file path the actual one to `CHARTEVENTS_DATA_TABLE.csv.gz` on your computer

```python
c_schema = Schema.rename(chartevents_schema, rename_cols={col: col.upper() for col in chartevents_schema.col_names()})
csv_loader = CsvDataStore(c_schema, '/Users/dan/dev/data/mimic3/CHARTEVENTS_DATA_TABLE.csv.gz', compress=True)
```

```python
chunks = csv_loader.load_chunks()
```

### Takes a long time to run.

```python
chartevents_store = CharteventsDataStore()

for chunk in chunks:
    print('appending a chunk')
    chunk.columns = [col.lower() for col in chunk.columns]
    chartevents_store.append(chunk)
```
